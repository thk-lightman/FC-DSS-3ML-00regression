```python
from sklearn.datasets import make_regression

X, y, w = make_regression(
            n_samples = 50, n_features = 1 , bias = 100, noise = 10, coef = True, random_state = 1)
```

```python
xx = np.linspace(-3, 3, 100)
y0 = w * xx + 100
plt.figure(figsize=(15,19))
plt.plot(xx,y0, "r-")
plt.scatter(X, y, s=100)
plt.xlabel("x")
plt.ylabel("y")
plt.title("make_regression ")
plt.show()
```

# [] make_regression2

```python
def make_regression2(n_samples, bias, noise, random_state=0):
    np.random.seed(random_state)
    X = np.random.normal(size = n_samples)
    w = 100 * np.random.normal(size=1)[0] # noise      
    y0 = w*X + bias
    e = np.random.normal(scale =noise, size = n_samples)
    y = y0 + e
    return X, y, w
```
